# Node Embeddings
- A class `NodeEmbeddings` that will return a concatenation of the selected features
- Features are based on shortest path generated by weights and distance between dummy source and sink nodes: 
    1. node embedding using (distance from source, distance to sink)
    2. distribution of 1-mers of the shortest path (distribution prefix, distribution suffix) 
    

```python
from panricci.utils.gfa_loader import GFALoader
from panricci.alignment.node_embeddings import NodeEmbeddings
```

```python
loader=GFALoader()
path = "/data/analysis-paper/experiments/mini-experiment-standard-not_alpha_consistent-not_fixblock/gfa-unchop/nodes/penalization0-min_len0-min_coverage0-alpha1/DQA1-3117.gfa"
path_chkpt = "../output/graph-alignment/ricci-flow/DQA1-3117-alpha1-ricciflow-5.edgelist"
G=loader(path, path_chkpt)
```

```python
node_embeddings = NodeEmbeddings(ricci_embedding=False, seq_embedding=True, kmer_size=2)
```

```python
embeddings = node_embeddings(G)
```
